```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt


X = df[['energy',	'tempo',	'valence',	'liveness']]
Y = df["duration"]

# Datos de entrenamiento y datos de prueba

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

X_train_constante = sm.add_constant(X_train)
X_test_constante = sm.add_constant(X_test)

modelo = sm.OLS(Y_train, X_train_constante).fit()


Yc = modelo.predict(X_test_constante)

plt.scatter(Y_test, Yc)

r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de correlacion: {r2: 0.4f}\n')
```

```python
b0, b1, b2, b3, b4 = modelo.params

Fun = lambda x1, x2, x3, x4: b0 + b1 * x1 + b2 * x2 + b3 * x3 + b4 * x4

# El salario estimado se puede calcular usando variables como el puntaje de crdito, edad, antigedad laboral y saldo financiero , ya que reflejan estabilidad, experiencia y capacidad de pago.
# Estos datos permiten predecir ingresos mediante modelos estadsticos o de aprendizaje automtico.

Fun(432, 60, 2, 120000)
```

```python
from statsmodels.formula.api import ols

moedo_2 = ols(formula = 'duration ~ energy + tempo + valence + liveness', data = df).fit()
tabla_anova = sm.stats.anova_lm(moedo_2, typ = 2)
tabla_anova
```

Energy: El valor P asociado a la energa (0.217564) es mayor que 0.05. Esto sugiere que, en este modelo y con estos datos, la energa no es un predictor estadsticamente significativo de la duracin de la cancin.

---



Tempo: El valor P asociado al tempo (0.104922) tambin es mayor que 0.05. Por lo tanto, el tempo no es un predictor estadsticamente significativo de la duracin de la cancin en este modelo.

---



Valence: El valor P asociado a la valencia (0.017078) es menor que 0.05. Esto indica que la valencia es un predictor estadsticamente significativo de la duracin de la cancin.

---



Liveness: El valor P asociado a la vivacidad (0.010405) es menor que 0.05. Esto indica que la vivacidad es un predictor estadsticamente significativo de la duracin de la cancin.

---



Residual: La fila "Residual" muestra la variabilidad que no es explicada por el modelo. Los valores NaN en las columnas F y PR(>F) son esperados para la fila residual.

---



La valencia y la vivacidad tienen una relacin estadsticamente significativa con la duracin de la cancin. La energa y el tempo no muestran una relacin estadsticamente significativa con la duracin de la cancin en este modelo particular. Es importante recordar que esta conclusin se basa en este conjunto de datos especfico y en el modelo lineal definido. Podran existir relaciones no lineales o interacciones entre las variables que no se capturan en este modelo simple.

# **Marco Teorico**

El presente estudio se basa en conceptos de la musicologa computacional y el anlisis de datos musicales. Se considera la influencia de las caractersticas intrnsecas de la msica en la respuesta del oyente y su adopcin en plataformas de streaming. Algunos conceptos clave que sustentan esta investigacin incluyen caractersticas de
- Audio Digital: Las plataformas de streaming como Spotify y otras analizan las canciones utilizando algoritmos para extraer caractersticas de audio cuantificables.
- Bailabilidad (Danceability): Una medida de qu tan adecuada es una cancin para bailar basada en elementos rtmicos, tempo, estabilidad del beat y fuerza general.
-Energa (Energy): Una medida de intensidad y actividad. Tpicamente, las canciones energticas se sienten rpidas, ruidosas y ruidosas.
-Tempo: La velocidad o ritmo de la msica, medida en pulsaciones por minuto (BPM).
-Volumen (Loudness): La cualidad subjetiva de la intensidad del sonido.
-Modo (Mode): Indica la modalidad (mayor o menor) de una pista, el tipo de escala de la que deriva su contenido meldico.
-Tonalidad (Key): La tonalidad global estimada de la cancin.
-Acstica (Acousticness): Una medida de qu tan acstica es la cancin.
-Instrumentalidad (Instrumentalness): Predice si una pista no contiene voces. Cuanto ms cerca est el valor a 1.0, mayor ser la probabilidad de que la pista no contenga contenido vocal.
-Vivacidad (Liveness): Detecta la presencia de una audiencia en la grabacin.
-Valencia (Valence): Una medida de la positividad musical transmitida por una pista.
-Duracin (Duration_ms): La duracin de la cancin en milisegundos.
-Speechiness: Detecta la presencia de palabras habladas en una pista.

```python
promedio = df['duration'].mean()
df['mayor_al_promedio'] = (df['duration'] > promedio).astype(int)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical features for correlation analysis
numerical_df = df.select_dtypes(include=['number'])

correlation_matrix = numerical_df.corr()
plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, cbar_kws={"shrink": .8})
plt.xlabel('Variables')
plt.ylabel('Variables')
plt.title('Mapa de Calor de la Matriz de Correlacin')
```

```python
#Seleccin de variables
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt

X = df[['duration', 'energy', 'tempo']]
Y = df['mayor_al_promedio']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Crear y entrenar el modelo
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, Y_train)

#Predecir
Y_pred = rf.predict(X_test)

#Evaluar
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

importancia = pd.Series(rf.feature_importances_, index=X.columns)
print(importancia.sort_values(ascending=False))
```

El informe de clasificacin indica que el modelo Random Forest tuvo un rendimiento ptimo con los datos de prueba. Logr un 100 % de precisin, recuperacin y puntuacin F1 para ambas clases, lo que result en una precisin general del 100 %. Esto sugiere que el modelo aprendi muy bien los patrones subyacentes en los datos y se generaliz perfectamente a los datos de prueba no analizados.

```python
from sklearn.tree import DecisionTreeClassifier

# Crear y entrenar el modelo
modelo_dtc = DecisionTreeClassifier(random_state=42)
modelo_dtc.fit(X_train, Y_train)

# Predecir
Y_pred = modelo_dtc.predict(X_test)

# Evaluar
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
```

El anlisis de importancia de las caractersticas revela que la "duracin" es la caracterstica ms importante (76,2%) para la prediccin, seguida por "tempo" (12,9%) y "energa" (11,0%), siendo esta ltima la menos influyente.
