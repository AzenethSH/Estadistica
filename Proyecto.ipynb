```python
import numpy as np

import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/AzenethSH/Estadistica/refs/heads/main/Experience-Salary.csv")
df.head()
X = df["exp(in months)"]
Y = df["salary(in thousands)"]

df.dropna(inplace = True)

#1.Establesca una variable dependiente ( Y ) y una variable independiente ( X )

#Variable independiente: examenes cortos
#Variable dependiente: salary(in thousands

#2.Realice un diagrama de dispersión para estos datos
import matplotlib.pyplot as plt
plt.scatter(X,Y, color="purple")
plt.xlabel('exp(in months)')
plt.ylabel('salary(in thousands')
plt.title('Grafico de dispersión')
plt.grid()

#3.¿Los datos soportan la suposición de linealidad?
# Si se puede trazar una linea

#4, Calcule el coeficiente de correlación e interprete el resultado.
Sxx = np.sum((X-np.mean(X))**2)
Sxy = np.sum((X-np.mean(X))*(Y-np.mean(Y)))
Syy = np.sum((Y-np.mean(Y))**2)
r = Sxy/np.sqrt(Sxx*Syy)
print("Coeficiente de correlación: ", r)

#5. Calcule el coeficiente de determinación e interprete el resultado.
r2 = r**2
print("Coeficiente de determinación: ", r2)
# que el 74.75% justifica la variabilidad de X y Y

#6. Obtenga la recta de regresión ajustada y grafíquelo sobre el gráfico de dispersión.
b1 = Sxy/Sxx
b0 = np.mean(Y) - b1*np.mean(X)
print(f"Pendiente: {b1}")
print(f"Intercepto: {b0}")

Yc = b1 * X + b0

import matplotlib.pyplot as plt
plt.sca se tter(X,Y, color="purple")
plt.plot(X,Yc, color="red")
plt.xlabel('exp(in months)')
plt.ylabel('salary(in thousands')
plt.title('Grafico de dispersión')

#7. Obtenga un intervalo de confianza del 95% para la pendiente de la recta de
#regresión ajustada ( b1 )
nivel_de_significancia = 0.05
import scipy.stats as st
t = st.t.ppf(1-nivel_de_significancia/2, len(X)-2)
se_b1 = np.sqrt(np.sum((Y-Yc)**2)/(len(Y)-2)) / np.sqrt(Sxx)
confianza_b1 = (b1-t*se_b1, b1+t*se_b1)
print(f"Intervalo de confianza para b1: {confianza_b1}")

#8. Calcule los residuales y trace un nuevo gráfico de dispersión. Comente,
#¿Parece que se verifican los supuestos?
residuales = Y - Yc
plt.figure()
plt.scatter(X,residuales, color="blue")
plt.xlabel("exp(in months)")
plt.ylabel("salary(in thousands")
plt.title("Grafico de residuales")
plt.axhline(y=0, color = "red", linestyle = "--")
plt.grid()

#9. Realice la prueba de Shapiro para los residuales y comente el resultado.
from scipy.stats import shapiro
_, valor_p_sh = shapiro(residuales)
print(f"valor_p de shapiro : {valor_p_sh}")
#No se descarta la hipotesis nula, los numeros si van bien


#11. Tres estudiantes sacaron 70, 75 y 84 de calificación. Según la recta de regresión
#  ajustada, ¿cuáles son los resultados esperados para estos tres alumnos?

print (b1 * 70 + b0)
print (b1 * 75 + b0)
print (b1 * 84 + b0)

#12. Realice una tabla ANOVA e interprete el resultado.

print (f"Sxx: {Sxx}")
print (f"Sxy: {Sxy}")
print (f"Syy: {Syy}")
print (f"SSE: {np.sum(residuales**2)}")
print (f"SST: {np.sum((Y-np.mean(Y))**2)}")
print (f"SSR: {np.sum((Yc-np.mean(Y))**2)/1}")
print (f"MRS: {np.sum((Yc-np.mean(Y))**2)/1}")
print (f"MSE: {np.sum(residuales**2)/(len(Y)-2)}")
print (f"F: {(np.sum((Yc-np.mean(Y))**2)/1)/(np.sum(residuales**2)/(len(Y)-2))}")
print (f"p-value: {1-st.f.cdf ((np.sum((Yc-np.mean(Y))**2)/1)/(np.sum(residuales**2)/(len(Y)-2)), 1, len(Y)-2)}")

```

#**Experience-Salary**


##*Introducción*

En el dinámico mundo laboral actual, la relación entre la experiencia laboral y el salario sigue siendo un tema de gran interés y relevancia. La creencia común es que a medida que un profesional adquiere más años de experiencia, su remuneración salarial tiende a aumentar. Sin embargo, esta relación no siempre es lineal y puede estar influenciada por diversos factores. Este estudio tiene como objetivo analizar la correlación entre la experiencia laboral y el salario, utilizando datos reales para proporcionar una visión empírica de esta relación. Específicamente, buscaremos determinar si existe una correlación positiva significativa entre la experiencia y el salario, y cómo esta relación se manifiesta en los datos proporcionados. A través de este análisis, esperamos ofrecer información valiosa tanto para profesionales que buscan mejorar su posición salarial como para empleadores que desean establecer políticas de compensación justas y equitativas.



#*Marco Teorico*

El análisis de la relación entre experiencia laboral y salario se fundamenta en diversas teorías económicas y laborales. La teoría del capital humano, por ejemplo, sugiere que la inversión en educación y experiencia aumenta la productividad de un individuo, lo que a su vez se traduce en un mayor salario. La experiencia laboral se considera una forma de acumulación de capital humano, ya que permite a los profesionales desarrollar habilidades y conocimientos específicos que son valorados por los empleadores. Por otro lado, la teoría de la señalización postula que la experiencia actúa como una señal para los empleadores, indicando la capacidad y competencia de un individuo. En este contexto, los años de experiencia pueden ser interpretados como un indicador de la calidad y el rendimiento potencial de un empleado. Además de estas teorías, es importante considerar otros factores que pueden influir en el salario, como el nivel educativo, las habilidades especializadas, el sector de la industria y la ubicación geográfica. En este estudio, nos centraremos en analizar la relación directa entre la experiencia y el salario, utilizando los datos proporcionados para identificar patrones y tendencias relevantes.

#*Hipótesis*

Basándonos en la teoría del capital humano y la creencia común sobre la relación entre experiencia y salario, formulamos la siguiente hipótesis principal:

Hipótesis Principal: Existe una correlación positiva significativa entre los años de experiencia laboral y el salario.
Además, podemos explorar hipótesis secundarias relacionadas con la fuerza de esta correlación y la posible existencia de valores atípicos o patrones no lineales en los datos.

#*Metodología*

Este estudio se basa en un análisis cuantitativo de los datos proporcionados en el archivo CSV "Experience-Salary.csv". La metodología incluye los siguientes pasos:



1.  **Recopilación de Datos**: Los datos se obtuvieron del archivo CSV proporcionado, que contiene información sobre los años de experiencia y los salarios de un grupo de profesionales.
2.   **Limpieza y Preparación de Datos**: Se realizó una revisión de los datos para identificar y corregir posibles errores o valores atípicos. Se verificó la integridad de los datos y se aseguraron de que estuvieran en un formato adecuado para el análisis.
3. **Análisis Descriptivo**: Se calcularon estadísticas descriptivas, como la media, la mediana, la desviación estándar y los percentiles, para comprender la distribución de los años de experiencia y los salarios.
4. **Análisis de Correlación**: Se utilizó el coeficiente de correlación de Pearson para medir la fuerza y la dirección de la relación lineal entre la experiencia y el salario.
5. **Análisis de Regresión**: Se realizó un análisis de regresión lineal para modelar la relación entre la experiencia y el salario, y para predecir el salario en función de los años de experiencia.
6. **Visualización de Datos**: Se crearon gráficos de dispersión y líneas de regresión para visualizar la relación entre la experiencia y el salario, y para identificar posibles patrones o tendencias.
7. **Software**: Para el análisis de los datos se podría utilizar Python con las librerías Pandas, NumPy, Matplotlib y Scikit-learn, o en este caso de Google Colab
